```python
from IPython.core.prefilter import PrefilterChecker
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
import plotly.express as px
import seaborn as sns

url = "https://datasource.kapsarc.org/explore/dataset/saudi-arabia-coronavirus-disease-covid-19-situation/download/?format=csv&timezone=Asia/Baghdad&lang=en&use_labels_for_header=true&csv_separator=%3B"
url = "https://datasource.kapsarc.org/explore/dataset/saudi-arabia-coronavirus-disease-covid-19-situation/download/?format=xls&timezone=Asia/Baghdad&lang=en&use_labels_for_header=true"

Covid = pd.read_csv ( url, header = None )

Name = [[" Daily / Cumulative" , " Indicator ", "Date", "Event", "region" , " City", " Cases"]] 

Covid.columus = Name

Covid.head(10) 
Covid.info()
Covid.describe() 
count = Covid["Cases" ].value_counts()
count.rename(index = {"Cases"  : "Value Count" }, inplace = True )
count.index.name = "Cases"
count
Covid.shape 

#drop all the rows with Null values
Covid = Covid.dropna() 
Covid.isna().sum()
Covid.isnull().sum() 
sum(Covid.duplicated()) # IF there is duplicate rows in the dataset

Covid = Covid.dropna(axis=0)
Covid.isna().sum()
 
Covid["Cases"] = Covid.loc[:, range(2, 1275)].sum(axis = 1) 

 #Drop the rows with column 'Total=0'
Covid = Covid[Covid['Total']!=0]
x = Covid
print(x)
Covid.to_csv("SACorona.csv")

#Missing Data 
Covid= Covid.inerpolate(method='linear')
Covid['Cases'].mean()
Covid1 = Covid.copy()
# Fill all NaN values with 0
Covid1['Cases'] = Covid1['Cases'].fillna(0)
Covid1['Cases'].mean()

#DATA TYPE
type(Covid)
print(Covid.dtypes)
Covid['Cases'].dtype
Covid["Cases"] = Covid ["Cases"].astype("int")
Covid.plot_id.astype("float")


Covid.drop(['Daily / Cumulative','Event'],axis=0 ,inplace=True)
# --

Covid["Cases"] = Covid["Cases"].replace("?" , np.nan)
Covid.dropna( subset = ["Cases"], axis = 0 , inplace = True )
Covid["Cases"] = Covid["Cases"].astype("int")
bins = np.linspace(Covid ["Cases"].min() , Covid ["Cases"].max(), 4)
Data = ["low", "mid", "high"]
Covid["Cases_bins"] = pd.cut(Covid["Cases"] , bins, labels = Data, include_lowest = True)
Covid["Cases_bins"]

Covid[[" Indicator " , "Cases"]].describe().round(2)

Covid.groupby([" Indicator " , "region" , "Cases "]).count()
gd = Covid.groupby([" Indicator " , "region" , "Cases "])
gd.first()
gd.get_group([" Indicator " , "region" , "Cases "])

Covid.plot()
x = Covid.Cases
y = Covid.Indicator
# if x =  y per weak 
   #  print(x)
plt.plot(x)
plt.figure(figsize=(15, 8 ))
plt.title("SA-Covid19")
plt.show()

Covid.set_index('city', inplace = True )
Covid.sort_values(['Cases'], ascending = False , axis = 0 , inplace = True )
Top5 = Covid.head()
Top5.plot(kind = "area")
plt.show

Covid['Total'] = Covid.sum(axis = 1)
Covid1 = Covid.groupby('Cases', axis = 0 ).sum()
plt.figure(figsize=(15, 8 ))
Covid1['Total'].plot(kind = 'pie')
plt.title("SA-Covid19")
plt.pie(Covid1)
plt.show()


! pip install folium
Covid_map = folium.Map()
Covid_map=folium.Map(location=[23.71, 44.12],zoom_start=10,control_scale=True)
Covid_map


```
